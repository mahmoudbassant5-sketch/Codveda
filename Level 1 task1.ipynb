The numerical columns (sepal_length, sepal_width, petal_length, and petal_width) were inspected and confirmed to already be stored in a consistent numeric format (float64).

```python
df.mode()
```

The mode was shown using the .mode() function in pandas, which identifies the most frequently occurring value in each column. The output shows that petal_length has two modes which are 1.4 and 1.5.

```python
df.hist(figsize=(10,6))
plt.suptitle("Distribution of Iris Features")
plt.show()
```

According to the histgrams above, the sepal length values concentrated between 5.0 and 6.5 cm. And that the Sepal width values were distributed between 2.7 and 3.5 cm. In the petal length, there a bimodal distribution which means two main clusters of values. Similarity

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="species", y="sepal_length", data=df)
plt.title("Sepal Length by Species")
plt.show()
```

```python
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", data=df)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
